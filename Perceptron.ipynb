# 라이브러리 가져오기

```
import numpy as np
import matplotlib.pyplot as plt
```

# 파라미터 초기화

```
lr = 0.01; # learning rate
T = 0.5; # threshold
hist_E=[]; # history of error
err = 100; # initialization of values
```

# 데이터 생성하고 그리기

```
n = 50; # data number

a = np.array([np.random.randn(n), np.random.randn(n)])
b = np.array([np.random.randn(n)+3, np.random.randn(n)+3])
X = np.concatenate((a,b), axis=1)

plt.scatter(X[0,:n], X[1, :n], c='r', marker=".")
plt.scatter(X[0, n:2*n], X[1, n:2*n], c='b', marker=".")
plt.show()
```

# 레이블 생성 및 가중치 초기화

```
y=np.concatenate((np.ones(n), np.zeros(n)),axis=0)
X=np.concatenate(([np.ones(n*2)],X), axis=0)
w = np.random.rand(X.shape[0])
```

# 에러가 0.1이하가 될 때까지 예측 및 학습 반복

```
# Gradient descent
while err>0.1:
    # prediction
    p = np.dot(w,X)
    out = np.where(p > T, 1, 0)
    # trainning
    cost= (1.0/2*n)* np.dot((out - y), X.T)
    w = w - lr*cost
    err=(1.0/(2*n))*np.sum((out-y)*(out-y))
    hist_E.append(err)
plt.plot(hist_E)
```
